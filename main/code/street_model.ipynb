```
from street_data_utils import *
clusters=200
cluster(file="GPS_Compass.csv", output="cluster.npy", n_clusters=clusters)

from load_fine_data import *
X_data,y_data  = load_data(dataset_dir="drive/My Drive/images",cluster_file="cluster.npy",coords_file="GPS_Compass.csv")

X_data = np.array(X_data)
y_data = np.array(y_data)
X_train,X_test,y_train,y_test = train_test_split(X_data, y_data, test_size=0.1)
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

del X_data
del y_data
print(len(X_train))
print(len(y_train))


from inception_v3 import *
model = InceptionV3(include_top=True, weights=None,classes=200)
sgd = SGD(lr=0.0001, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(optimizer=sgd, loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=150)
```
