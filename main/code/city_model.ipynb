```
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten,MaxPooling2D
from load_data import *
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
from google.colab import drive
from keras.optimizers import SGD

drive.mount('/content/drive', force_remount=True)

    

X_data,y_data  = load_data("drive/My Drive/imgs")


X_train,X_test,y_train,y_test = train_test_split(X_data, y_data, test_size=0.15, stratify = y_data)
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)







```

```
from inception_v3 import *
model = InceptionV3(include_top=True, weights=None,classes=4)
sgd = SGD(lr=0.0001, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(optimizer=sgd, loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=150)
```

```
model.save("drive/My Drive/Models/gognet_100.hdf5")
```

```
model = Sequential()

#add model layers

model.add(Conv2D(32,kernel_size=5,strides=(1,1),activation='relu',input_shape=(256,256,3)))

model.add(MaxPooling2D(pool_size=(3,3),strides=(2,2)))
model.add(Conv2D(filters=64,kernel_size=5,strides=(1,1),activation='relu'))
model.add(MaxPooling2D(pool_size=(3,3),strides=(2,2)))
model.add(Conv2D(filters=64,kernel_size=5,strides=(1,1),activation='relu'))
model.add(MaxPooling2D(pool_size=(3,3),strides=(2,2)))
model.add(Flatten())
model.add(Dense(units=4,activation="softmax"))

#model = VGG_16()
sgd = SGD(lr=0.00001, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(optimizer=sgd, loss='categorical_crossentropy', metrics=['accuracy'])

#model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=50)

```
